## Use melt to re-structure the absolute values first. 

```python
df.columns
```

```python
df_per_serving = df[['chain', 'category_clean', 'item_name', 'serve_size', 'serve_size_unit',
       'calories_kcal', 'total_fat_g', 'saturated_fat_g', 'trans_fat_g',
       'cholesterol_mg', 'sodium_mg', 'carbohydrates_g', 'dietary_fiber_g',
       'sugars_g', 'protein_g']]
```

```python
col_stay = ['chain', 'category_clean', 'item_name', 'serve_size', 'serve_size_unit']
```

```python
df_per_serving = (pd.melt(df_per_serving, 
                    id_vars = col_stay,
                    value_name = "value_per_serving",
                    var_name = "nutrition_indicator")
                    .sort_values(by=["chain", "category_clean", "item_name"]))
```
