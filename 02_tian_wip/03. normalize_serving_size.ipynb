## Converting the serving size from fl oz to ml

```python
df[col] = df[col].apply(pd.to_numeric, errors='coerce')
df
```

```python
oz_t_ml = 28.413075
```

```python
df_ml = df.loc[df["serve_size_unit"].str.contains("fl oz")]
df_g = df.loc[df["serve_size_unit"].str.contains("g")]
```

```python
df_ml = df_ml.assign(serve_size = df_ml["serve_size_new"].mul(oz_t_ml))
df_ml["serve_size_unit"] = "ml"
```

```python
df_g = df_g.assign(serve_size = df_g["serve_size_new"].copy())
df_g
```

```python

df_combined = pd.DataFrame()
df_combined = df_combined.append(df_g).append(df_ml)
df_combined.head()
```
