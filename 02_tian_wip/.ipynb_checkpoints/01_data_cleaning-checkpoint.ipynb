### Harmonize the category of all 3 resturants

#### Mcdonalds

```python
mcdonalds["category"].value_counts()
```

```python
mcd_cat_dict = {"Coffee & Tea" : "beverage",
                "Breakfast" : "breakfast",
                "Smoothies & Shakes" : "beverage",
                "Chicken & Fish" : "burger",
                "Beverages" : "beverage",
                "Beef & Pork" : "burger",
                "Snack & Sides" : "sides",
                "Desserts" : "dessert",
                "Salads" : "salad"}
```

```python
mcdonalds = mcdonalds.assign(category_clean = "NaN")
mcdonalds["category_clean"] = mcdonalds["category"].map(mcd_cat_dict).fillna("no match")
mcdonalds["category_clean"].value_counts()
```

```python
mcdonalds.loc[mcdonalds["category_clean"] == "no match"]
```

```python
mcdonalds.iloc[90:96,-1] = "wrap"
```

```python
mcdonalds.iloc[96:99,-1] = "sides"
```

```python
mcdonalds.iloc[99:100,-1] = "kids_menu"
```

```python
mcdonalds.iloc[100:101,-1] = "salad"
```

```python
mcdonalds.iloc[101:103,-1] = "dessert"
```

```python
subway_cat_old = list(subway["category"].value_counts().to_dict().keys())
subway_cat_old
```

```python
subway_cat_new = ["sandwich", "extra", "wrap", "bread", "dessert", "sauce", "breakfast", "salad"]
```

```python
subway_dict = dict(zip(subway_cat_old, subway_cat_new))
```

```python
subway = subway.assign(category_clean = "NaN")
```

```python
subway["category_clean"] = subway["category"].map(subway_dict).fillna("no match")
```
