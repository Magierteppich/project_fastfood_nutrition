## 3D Scatterplot with plotly

```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import warnings
warnings.filterwarnings('ignore')
```

```python
# putting the data into a dataframe

menu = pd.read_csv('00_comibined_final.csv')
menu.head(2)
```

```python
# Check for Nulls
print(menu.isnull().any())
print("--------------------------------------")
# check for numbers
print(menu.describe())
print("--------------------------------------")
```

```python
# 3D scatter plot for calories

trace1 = go.Scatter3d(
    x=menu['category_clean'].values,
    y=menu['item_name'].values,
    z=menu['calories_kcal'].values,
    text=menu['item_name'].values,
    mode='markers',
    marker=dict(
        sizemode='diameter',
#         sizeref=750,
#         size= dailyValue['Cholesterol (% Daily Value)'].values,
        color = menu['calories_kcal'].values,
        colorscale = 'Portland',
        colorbar = dict(title = 'calories_kcal'),
        line=dict(color='rgb(255, 255, 255)')
    )
)

data=[trace1]
layout=dict(height=800, width=800, title='3D Scatter Plot of calories_kcal')
fig=dict(data=data, layout=layout)
py.iplot(fig, filename='3DBubble')
```
